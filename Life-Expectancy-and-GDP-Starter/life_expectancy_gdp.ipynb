```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("GDP_LifeExpectancy.csv")
```

```python
print(df.head())
print(df.info())
print(df.describe(include = "all"))
print(df.Country.unique())
```

```python
#overall
sns.scatterplot(data = df, x = "GDP", y = "Life expectancy at birth (years)")
plt.show()
```

```python
#by country
sns.scatterplot(data = df, x = "GDP", y = "Life expectancy at birth (years)", hue = "Country")
plt.show()
```

```python
#by year
sns.scatterplot(data = df, x = "GDP", y = "Life expectancy at birth (years)", hue = "Year")
plt.show()
```

```python
#overall with country and year
sns.scatterplot(data = df, x = "GDP", y = "Life expectancy at birth (years)", hue = "Year", style = "Country", s = 100)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1)
plt.show()
```

```python
fig, axes = plt.subplots(2,3, figsize = (15,10))

country = df.Country.unique()

for i in range(len(country)):
    if i < 3:
        sns.scatterplot(ax = axes[0,i], data = df[df.Country == country[i]], x = "GDP", y = "Life expectancy at birth (years)", 
                        hue = "Year", legend = False)
        axes[0,i].set_title(country[i])
    else:
        n = i-3
        sns.scatterplot(ax = axes[1,n], data = df[df.Country == country[i]], x = "GDP", y = "Life expectancy at birth (years)", 
                        hue = "Year", legend = False)
        axes[1,n].set_title(country[i])

fig.suptitle("Life Expectancy at Birth and GDP per Country", fontsize = 25)
plt.show()

```

```python
fig, axes = plt.subplots(2,3, figsize = (15,10))

country = df.Country.unique()

for i in range(len(country)):
    if i < 3:
        sns.regplot(ax = axes[0,i],data = df[df.Country == country[i]], x= "GDP", y = "Life expectancy at birth (years)" )
        axes[0,i].set_title(country[i])
    else:
        n = i-3
        sns.regplot(ax = axes[1,n],data = df[df.Country == country[i]], x= "GDP", y = "Life expectancy at birth (years)" )
        axes[1,n].set_title(country[i])

fig.suptitle("Life Expectancy at Birth and GDP per Country", fontsize = 25)
plt.show()

```
